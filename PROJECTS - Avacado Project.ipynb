# Treating Outliers

```python
df.plot(kind='box',subplots=True,layout=(5,7),figsize=(16,24)) 
plt.show()
```

From above Boxplots we can see that Outliers are present in many columns like :- 'Total Volume','4046','4225','4770','Total Bags','Small Bags','Large Bags' & 'XLarge Bags'.

Splitting data into Target & Independent columns

```python
x=df.iloc[:,1:]     #Independent Columns

y=pd.DataFrame(df.iloc[:,0])   #Target Column
```

Removing Outliers using Z-score

```python
from scipy.stats import zscore               # Removing Outliers from independent variables
z=np.abs(zscore(x))
z
```

```python
# Now to take out the ouliers taking threshold 3
# All values with Z-score >3 will be considered as ouliers
# Making a new file for values with z-score<3

threshold=3
x1=x[(z<3).all(axis=1)]
x1
```

```python
print(np.where(z>3)[0])
```

```python
y1= y.drop(np.where(z>3)[0],axis=0)     # Dropping the rows from same locations from target varibles
y1
```

```python
x1.shape
```

# Treating Skewness

```python
x1.skew().sort_values(ascending=False)
```

Taking Skewness threshold +/- 0.5. We can see that the following columns are having skewnees which needs to be removed:-
   1. XLarge Bags
   2. 4770
   3. Large Bags
   4. 4046
   5. 4225
   6. Small Bags
   7. Total Bags
   8. Total Volume

All, these above columns are Right Skewed.

Using Power Transformation for removing skewnees from the data.

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson',standardize=True)
```

```python
a = x1.loc[:,['XLarge Bags','4770','Large Bags','4046','4225','Small Bags','Total Bags','Total Volume']]    
a.head()                     # Creating Dataframe of the columns that need to be treated
```

```python
a.skew()
```

```python
data = pt.fit_transform(a)
a_new = pd.DataFrame(data,columns=a.columns,index=a.index)
a_new.skew()
```

Now we can see that the Skewnees is removed. Saving this transformed data to the original DataFrame. 

```python
x1['XLarge Bags']=a_new['XLarge Bags']
x1['4770']=a_new['4770']
x1['Large Bags']=a_new['Large Bags']
x1['4046']=a_new['4046']
x1['4225']=a_new['4225']
x1['Small Bags']=a_new['Small Bags']
x1['Total Bags']=a_new['Total Bags']
x1['Total Volume']=a_new['Total Volume']
x1.skew()
```

# Treating Multicollinearity

Checking Variance Inflation Factor(VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calc_vif(x):
    vif=pd.DataFrame()
    vif["Variables"]=x.columns
    vif["VIF FACTOR"]=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return(vif)
```

```python
calc_vif(x1)
```

From above we can see that Total Volumne & Total Bags are having Highest V.I.F. Values close to each other. Removing these columns one by one.

```python
x1.drop({'Total Volume'},axis=1,inplace=True)            # Total Volume 
calc_vif(x1)
```

```python
x1.drop({'Total Bags'},axis=1,inplace=True)  
calc_vif(x1)
```

Now we can see that V.I.F. of all the Columns have been reduced below 10. So, moving ahead.

# Data Scaling

As we can see the columns are in different scales. So we will use encoding techniques to scale the data.

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
```

```python
scale=st.fit_transform(x1)
x1=pd.DataFrame(scale,columns=x1.columns)
x1
```

Now we can see all the Input data is in same scale.

# Splitting the Data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression 
lm = LinearRegression()
```

Finding out best random state.

```python
for i in range(0,200):
    x1_train,x1_test,y1_train,y1_test= train_test_split(x1,y1,random_state=i,test_size=0.22)
    lm.fit(x1_train,y1_train)
    pred_train = lm.predict(x1_train)
    pred_test = lm.predict(x1_test)
    if round(r2_score(y1_train,pred_train)*100,1) == round(r2_score(y1_test,pred_test)*100,1):
        print("At random state",i,"the Model performs very well")
        print(f"For {i},the training accuracy is :- {r2_score(y1_train,pred_train)*100}")
        print(f"For {i},the testing accuracy is :- {r2_score(y1_test,pred_test)*100}")
        print(f"For {i},the Mean squared error is :- {mean_squared_error(y1_test,pred_test)}")
        print("\n")
```

From above we can see that Model is performing best at Random State = 144. Beacuse for random state=144 Model's testing accuracy is equal to Training accuracy. And Model is getting least mean squared error.

Therefore, using Random State = 144 for further working & Splitting the DataFrame.

```python
x1_train,x1_test,y1_train,y1_test= train_test_split(x1,y1,random_state=144,test_size=0.22)
```

# Testing with Different Models.

```python
from sklearn.linear_model import LinearRegression 
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
lm = LinearRegression()
svr = SVR(kernel='rbf')
dtc = DecisionTreeRegressor()
ls  = Lasso(alpha=0.0001)
rd = Ridge(alpha=0.0001)

models = [lm,svr,dtc,ls,rd]

for i in models:
    i.fit(x1_train,y1_train)
    pred_train = i.predict(x1_train)
    pred_test = i.predict(x1_test)
    print(f"For {i},the training accuracy is :- {r2_score(y1_train,pred_train)*100}")
    print(f"For {i},the testing accuracy is :- {r2_score(y1_test,pred_test)*100}")
    print(f"For {i},the Mean squared error is :- {mean_squared_error(y1_test,pred_test)}")
    print("\n")
```

1. From above we can see that SVR() is working best with the data with highest accuracies & with least Mean squared error. 
2. DecisionTreeRegressor() is also performing well with the dataset.

# Ensemble Techniques

Using various Bagging & Boosting Ensemble Algorithms to increase the model performance.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100)
br = BaggingRegressor(n_estimators=100)
ad = AdaBoostRegressor(n_estimators=100)
gb = GradientBoostingRegressor(n_estimators=100)

models = [rf,br,ad,gb]

for i in models:
    i.fit(x1_train,y1_train)
    pred_train = i.predict(x1_train)
    pred_test = i.predict(x1_test)
    print(f"For {i},the training accuracy is :- {r2_score(y1_train,pred_train)*100}")
    print(f"For {i},the testing accuracy is :- {r2_score(y1_test,pred_test)*100}")
    print(f"For {i},the Mean squared error is :- {mean_squared_error(y1_test,pred_test)}")
    print("\n")
```

From above ensemble methods we can see that RandomForestClassifier & BaggingRegressor are working best with the Data.

Whereas out of these two, BaggingRegressor is performing best beacuse the Mean Sqaured Error in less.

# Cross-Validation of Algorithms

Checking cross_val_scores for all the Algorithms that are performing well to know the which Algorithm is actually working best for the the Model.

```python
from sklearn.model_selection import cross_val_score
model = [svr,dtc,rf,br]

for m in model:
    pred_test = m.predict(x1_test)
    score = cross_val_score(m,x1,y1,cv=9)
    c_score = score.mean()
    print("For",m,"Cross Validation score is:-",c_score*100)
    print(f"The testing accuracy is :- {r2_score(y1_test,pred_test)*100}")
    print("\n")
```

Now from above we can confirm that BaggingRegressor(n_estimators=100) is working Best for the data beacuse:-

1. BaggingRegressor is getting Highest accuracy score among all the algorithms.
2. Cross Val-Score is highest is case of BaggingRegressor.
3. Mean squared error is minimum in case of BaggingRegressor.

Using Lasso as the Final Algorithm & Optimizaztion.

# Cross-Validation of Algorithms

Checking cross_val_scores for all the Algorithms that are performing well to know the which Algorithm is actually working best for the the Model.

```python
from sklearn.model_selection import cross_val_score
model = [svc,dtc,rf,bg]

for m in model:
    pred_test=m.predict(x2_test)
    score = cross_val_score(m,x2,y2,cv=5)
    c_score = score.mean()
    print("For",m,"Cross Validation score is:-",c_score*100)
    print('Accuracy Score of',m,'is:',accuracy_score(y2_test,pred_test)*100)
    print("\n")
```

Now from above we can confirm that RandomForestClassifier() is working Best for the data beacuse:-

1. RandomForestClassifier() is getting Highest accuracy score among all the algorithms.
2. Cross Val-Score is highest is case of RandomForestClassifier().

Therefore, Using RandomForestClassifier() as the Final Algorithm & Optimizaztion.
