# Treating Multicollinearity

Checking Variance Inflation Factor(VIF)

Splitting data into Target & Independent columns.

```python
x=df.iloc[:,1:]

y=pd.DataFrame(df.iloc[:,0])
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(x):
    vif=pd.DataFrame()
    vif["Variables"]=x.columns
    vif["VIF FACTOR"]=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return(vif)
```

```python
calc_vif(x)
```

For above V.I.F. Values we can see that 
1. ER is having highest V.I.F. Value. We can drop this column.
2. ERA & RA also having very high V.I.F. close to each other. We can drop 'RA' column beacuse ERA is more correlated with the W(target.)
3. AB has the next highest range of V.I.F. so we can drop this columnn too.

```python
x.drop({'ER'},axis=1,inplace=True)                     # Dropping ER column
calc_vif(x)
```

```python
x.drop({'RA'},axis=1,inplace=True)                  # Dropping RA column which has highest V.I.F.
calc_vif(x)
```

```python
x.drop({'AB'},axis=1,inplace=True)                  # Dropping AB column which has highest V.I.F.
calc_vif(x)
```

```python
x.drop({'H'},axis=1,inplace=True)                  # Dropping H column which is having next higest V.I.F
calc_vif(x)
```

```python
x.drop({'2B'},axis=1,inplace=True)                  # Dropping 2B column instead of R as R shows more correlation with W   
calc_vif(x)
```

Now we can see that the V.I.F. Values have been reduced significantly. So, proceeding further.

# Splitting the Data

Finding out best random state.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression 
lm = LinearRegression()
```

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=i,test_size=0.22)
    lm.fit(x_train,y_train)
    pred_train = lm.predict(x_train)
    pred_test = lm.predict(x_test)
    if round(r2_score(y_train,pred_train)*100,1) == round(r2_score(y_test,pred_test)*100,1):
        print("At random state",i,"the Model performs very well")
        print(f"For {i},the training accuracy is :- {r2_score(y_train,pred_train)*100}")
        print(f"For {i},the testing accuracy is :- {r2_score(y_test,pred_test)*100}")
        print(f"For {i},the Mean squared error is :- {mean_squared_error(y_test,pred_test)}")
        print("\n")
```

From above we can see that Model is performing best at Random State = 197. Beacuse for random state=197 Model's testing accuracy is equal to Training accuracy. And Model is getting Highest accuracy with less mean squared error

Therefore, using Random State = 197 for further working & Splitting the DataFrame

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=197,test_size=0.22)
```

# Testing with Different Models

```python
from sklearn.linear_model import LinearRegression 
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
lm = LinearRegression()
svr = SVR(kernel='rbf')
dtc = DecisionTreeRegressor()
ls  = Lasso(alpha=0.0001)
rd = Ridge(alpha=0.0001)

models = [lm,svr,dtc,ls,rd]

for i in models:
    i.fit(x_train,y_train)
    pred_train = i.predict(x_train)
    pred_test = i.predict(x_test)
    print(f"For {i},the training accuracy is :- {r2_score(y_train,pred_train)*100}")
    print(f"For {i},the testing accuracy is :- {r2_score(y_test,pred_test)*100}")
    print(f"For {i},the Mean squared error is :- {mean_squared_error(y_test,pred_test)}")
    print("\n")
```

From above we can see that LinearRegression(), Lasso(alpha=0.0001) & Ridge(alpha=0.0001) are working good with the data and are getting high accuracies. Whereas b/w these three Lasso is working best because the Mean Sqaured Error is minimum in this case. 

# Ensemble Techniques

Using various Bagging & Boosting Ensemble Algorithms to increase the model performance.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100,random_state=42)
br = BaggingRegressor(n_estimators=100,random_state=42)
ad = AdaBoostRegressor(n_estimators=100)
gb = GradientBoostingRegressor(n_estimators=100)

models = [rf,br,ad,gb]

for i in models:
    i.fit(x_train,y_train)
    pred_train = i.predict(x_train)
    pred_test = i.predict(x_test)
    print(f"For {i},the training accuracy is :- {r2_score(y_train,pred_train)*100}")
    print(f"For {i},the testing accuracy is :- {r2_score(y_test,pred_test)*100}")
    print(f"For {i},the Mean squared error is :- {mean_squared_error(y_test,pred_test)}")
    print("\n")
```

Ensemble Techniques are not working well will the data as the data is small.

From above ensemble methods we can see that RandomForestClassifier,BaggingRegressor, AdaBoostRegressor & GradientBoostingRegressor() are obatining very high training accuracies but its difference to the Testing Accuracies are very high. AlSo, Mean Sqauared Error is also very high in case of Essemble Methods.

So these ensemble techniques cannot be used here.

# Cross-Validation of Algorithms

Checking cross_val_scores for all the Algorithms to know the which Algorithm is actually performing best for the the Model.

```python
from sklearn.model_selection import cross_val_score
model = [lm,ls,rd,rf,br,ad,gb,sgd]

for m in model:
    pred_test = m.predict(x_test)
    score = cross_val_score(m,x,y,cv=4)
    c_score = score.mean()
    print("For",m,"Cross Validation score is:-",c_score*100)
    print(f"The testing accuracy is :- {r2_score(y_test,pred_test)*100}")
    print("\n")
```

Now from above we can confirm that Lasso(alpha=0.0001) is working Best for the data beacuse:-
1. Lasso is getting Highest score among all the algorithms.
2. Cross Val-Score is highest is case of Lasso.
3. Mean squared error is minimum in case of Lasso.

Using Lasso as the Final Algorithm & Optimizaztion.
